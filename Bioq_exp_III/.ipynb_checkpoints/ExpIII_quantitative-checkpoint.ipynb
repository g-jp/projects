```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
from collections import Counter

BYall = pd.read_csv('BYG1.tsv', sep = '\t')

BYallmet_one = BYall.drop_duplicates(subset='KEGG_cid',keep='first').set_index('KEGG_cid')
BYallmet = BYall.set_index('KEGG_cid')


KOall = pd.read_csv('KOG1.tsv', sep = '\t')

KOallmet_one  = KOall.drop_duplicates(subset='KEGG_cid',keep='first').set_index('KEGG_cid')
KOallmet = KOall.set_index('KEGG_cid')

cid_list = []
for n in BYall.KEGG_cid:
    cid = n
    cid_list.append(cid)

```

```python
def bar_plot(x):
    val = []
    name = []
    a = KOallmet.peak_height[str(x)].sum()
    b = BYallmet.peak_height[str(x)].sum()
    enk = KOallmet.peak_height.HMDB01045.sum()
    qnt_a = a/enk
    qnt_b = b/enk
    val.append(qnt_a)
    val.append(qnt_b)
    name.append(KOallmet_one.KEGG_name[str(x)].split(';')[0] + '\nKO')
    name.append(BYallmet_one.KEGG_name[str(x)].split(';')[0] + '\n BY')

    name_1 = BYallmet_one.KEGG_name[str(x)].split(';')[0]
    with sns.axes_style('whitegrid'):
        plt.subplots(figsize=(3,5))
        c = sns.barplot(val,name)
    return c and print(f'The normalization for {name_1} is {val}')
```

```python
def norm_name(x):
    name = []
    a = KOallmet_one.KEGG_name[str(x)].split(';')[0] 
    b = BYallmet_one.KEGG_name[str(x)].split(';')[0] 
    if a[-1] == ')':
        a = a[:-9] + '\nKO'
        b = b[:-9] + '\nBY'
    else:
        a = a + '\nKO'
        b = b + '\nBY'
    
    name.append(a)
    name.append(b)


    return name

def norm_val(x):
    val = []
    heighta = KOallmet.peak_height[str(x)].sum()
    heightb = BYallmet.peak_height[str(x)].sum()
    enk = KOallmet.peak_height.HMDB01045.sum()
    qnt_a = heighta/enk
    qnt_b = heightb/enk
    val.append(qnt_a)
    val.append(qnt_b)
    
   
    return val

def plot_cids(x):

    ord = []
    for n in x:
        a = norm_val(n)
        for y in a:
            ord.append(y)    
    abc = []
    for n in x:
        a = norm_name(n)
        for y in a:
            abc.append(y)
    with sns.axes_style('whitegrid'):
        plt.subplots(figsize=(5,5))
        c = sns.barplot(ord,abc)
    return c
```
